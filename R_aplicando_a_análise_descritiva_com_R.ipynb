```r
dados_limpos <- na.omit(dados)
```

```r
sum(is.na(dados_limpos))
```

```r
colunas <- c("area", "rent", "total")

for(coluna in colunas){
  resultado <- any(dados_limpos[[coluna]] == 0)
  print(paste('A', coluna, 'tem numero 0?', resultado))
}
```

```r
dados_limpos %>% dplyr::filter(area == 0)
```

```r
dados_limpos <- dados_limpos %>% dplyr::filter(area != 0)
```

```r
dados_limpos %>% dplyr::filter(area == 0)
```

```r
valor_por_mq <- dados_limpos[, 'rent'] / dados_limpos [, 'area']
```

```r
head(valor_por_mq)
```

```r
dados_limpos %>% mutate(valor_por_mq)
```

```r
total_por_mq <- dados_limpos[, 'total'] / dados_limpos [, 'area']
```

```r
dados_limpos %>% mutate(total_por_mq)
```

```r
dados_limpos <- dados_limpos %>% mutate(
  valor_por_mq,
  total_por_mq
)
```

```r
dados_limpos %>% head()
```

```r
aggregate(list(Aluguel = dados_limpos$rent), list(Tipo = dados_limpos$type), mean)
```

```r
aggregate(dados_limpos %>% select(rent, total, valor_por_mq, total_por_mq), list(Tipo = dados_limpos$type), mean)
```

```r
aggregate(dados_limpos %>% select(rent, total, valor_por_mq, total_por_mq), list(Tipo = dados_limpos$type), median)
```

```r
aggregate(dados_limpos %>% select(rent, total, valor_por_mq, total_por_mq), list(Tipo = dados_limpos$type), sd)
```

```r
options (repr.plot.width = 10, repr.plot.height = 8)

boxplot(
  dados_limpos$total,
  horizontal = TRUE,
  main = 'Bloplot dos Alugueis com Taixa',
  lab = 'Aluguel (R$)',
  col = 'lightblue',
  outline = FALSE
)
```

```r
boxplot(
    total ~ type,
    data = dados_limpos,
    horizontal = FALSE,
    main = "Boxplot dos Alugueis com Taxas",
    xlab = "Tipo de imóvel",
    ylab = "Aluguel (R$)",
    col = 'red',
    las = 1,
    outline = FALSE
)
```

```r
ggplot(
  dados_limpos,
  aes(x = type, fill = type)
) + geom_bar () + theme_classic() +
  labs(
    title = 'Contagem de Imóveis por Tipo',
    x = 'Tipo de Imóvel',
    y = 'Quantidade de Imóveis'
  ) +
    theme(
    legend.position = 'none',
    plot.title = element_text(hjust = 0.5, size = 18)
    ) +
      geom_text (
      stat = 'count',
      aes(label = after_stat(count)),
      vjust = -0.5
      )
```

```r
ggplot(
  dados_limpos,
  aes(x = bedrooms)
) + geom_bar () + theme_classic() +
  labs(
    title = 'Contagem de Total de Quartos',
    x = 'Total de Quartos',
    y = 'Quantidade de Imóveis'
  ) +
    theme(
    legend.position = 'none',
    plot.title = element_text(hjust = 0.5, size = 18)
    ) +
      geom_text (
      stat = 'count',
      aes(label = after_stat(count)),
      vjust = -0.5
      )
```

```r
ggplot(dados_limpos, aes(x = area, y = total, color = type)) +
  geom_point(size = 4) +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  theme_minimal() +
  labs(
  title = 'Contagem de Valor por M²',
  x = 'Area em m²',
  y = 'Aluguel + Despesas (R$)') +
    theme(
    legend.position = 'bottom',
    plot.title = element_text (hjust = 0.5, size = 18))
```

```r
dados_numericos <- dados_limpos %>% select_if(is.numeric)
matriz_correlacao <- cor(dados_numericos, use = 'complete.obs')
matriz_correlacao
```

```r
modelo <- lm(total ~ area + bedrooms + garage, data = dados_limpos)

summary(modelo)
```

```r
plot(modelo, which = 1)
```

```r
plot(modelo, which = 2)
```

```r
valores_ajustados <- fitted(modelo)

ggplot(data = dados_limpos, aes(x = valores_ajustados, y = total)) +
  geom_point() +
  geom_abline(slope = 1, intercept = 0, col = "red", linetype = "dashed") +
  labs(title = "Valores Ajustados vs. Valores Observados",
       x = "Valores Ajustados", y = "Valores Observados") +
  theme_minimal()
```
